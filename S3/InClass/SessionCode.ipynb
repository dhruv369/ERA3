```python

from torch.utils.data import DataLoader, dataset
from matplotlib.pyplot import imshow
from torchvision import datasets, transforms, utils
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
class CustomDataset(dataset):
    def __init__(self, data, target):
        self.data = data
        self.target = target

    def __getitem__(self, index):
        x = self.data[index]
        y = self.target[index]
        return x, y

    def __len__(self):
        return len(self.data)
```

```python
transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transforms)
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
```

```python
dataiter = iter(train_loader)
images, labels = next(dataiter)
print(images.shape)
print(labels.shape)
```

```python
def imshow(img):
    img = img / 2 + 0.5 # unnormalize the image (reverse of normalization) 
    npimg = img.numpy() # convert the tensor to numpy for displaying
    plt.imshow(np.transpose(npimg, (1, 2, 0))) # transpose the image to (height, width, channels)
    plt.show()  # display it
    
```

```python
imshow(utils.make_grid(images))
print(' '.join('%5s' % labels[j] for j in range(4)))
```
